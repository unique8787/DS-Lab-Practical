```python

# Load dataset
df = pd.read_csv("marks_sem1.csv")

# Identify numeric columns
numeric_cols = df.select_dtypes(include='number').columns
print("Numeric Columns:", list(numeric_cols))

# Loop through numeric fields
for col in numeric_cols:
    data = df[col].dropna()

    print("\n====================================")
    print(f"Checking distribution for: {col}")
    print("====================================")

    # Histogram
    plt.figure()
    plt.hist(data, bins=10)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

    # KDE
    plt.figure()
    data.plot(kind='kde')
    plt.title(f"KDE of {col}")
    plt.xlabel(col)
    plt.ylabel("Density")
    plt.show()

    # Qâ€“Q Plot
    plt.figure()
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(f"Qâ€“Q Plot for {col} (Normal Distribution Check)")
    plt.show()

    # Skewness
    skew = data.skew()
```

```python
# Select numeric columns *as a DataFrame*
numeric_df = df.select_dtypes(include='number')

# Summary statistics
summary = numeric_df.describe().T
summary["skewness"] = numeric_df.skew()
summary["kurtosis"] = numeric_df.kurt()

summary
```
